# Predicting customer satisfaction via ML

The client is a startup in the logistics and delivery domain, whose main goal is their customers' satisfaction. Getting feedback from customers is not always easy, but it is the one effective way to gauge customers' satisfaction, and improve their operations accordingly. The company provides us with a subset of a bigger survey, and asked to come up with the most effective ML method to predict customers' happiness from their answers to the survey. 

In particular, the company asked to:

1. create a ML model that is at least 73% accuracy in predicting customer's satisfaction;
2. understand which questions are the most crucial to make correct predictions, and which can be removed from the survey without impinging on model accuracy.

## Libraries

```python
# basics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
#from xgboost import XGBClassifier # the sklearn GradientBoosting models are reported less performative than the xgboost ones
from sklearn.svm import SVC #support vector classifier

# accuracy metrics
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score, cross_validate, GridSearchCV, train_test_split
```

## Data exploration

```python
data = pd.read_csv("../data/raw/ACME-HappinessSurvey2020.csv")
data.head()
```

Data Description:

- Y = target attribute (Y) with values indicating 0 (unhappy) and 1 (happy) customers
- X1 = my order was delivered on time
- X2 = contents of my order was as I expected
- X3 = I ordered everything I wanted to order
- X4 = I paid a good price for my order
- X5 = I am satisfied with my courier
- X6 = the app makes ordering easy for me

Attributes X1 to X6 indicate the responses for each question and have values from 1 to 5 where the smaller number indicates less and the higher number indicates more towards the answer.

```python
data['Y'].value_counts().plot.bar()
```

From the plot above, we can see that there were more happy customers than unhappy customers (though the difference between the two is not as big as one may want to).

```python
sns.pairplot(data, corner=True, kind='reg')
```

```python
correlations = data.corr()
sns.heatmap(correlations, annot=True)
```

As we can see from the correlation heatmap above, it looks like customer happiness correlates more with X1 ('my order was delivered on time') and X5 ('I am satisfied with my courier') than with any other feature. This seems to suggest that customery satisfaction hinges on the experience with courier more than anything else (i.e., order content, price, and app experience).

Notably, X1 ('my order was delivered on time') highly correlates with and X3 ('I ordered everything I wanted to order'), X5 ('I am satisfied with my courier') and X6 ('the app makes ordering easy for me'). All other combinations had lesser correlation coefficients (< 0.1).

Therefore, it looks like X2 ('The content of my order was as I expected') and X4 ('I paid a good price for my order') are not so informative, and may be removed from analysis.

## Modelling

First, let's split the data so that the features *Xn* are separated from the to-be-predicted feature *y* (customer happiness). Then, we split the dataset in two subsets: the _train_ set of *Xn* and *y* features will be used to identify the right ML algorithm to predict customer happiness; the _test_ set will be used to evaluate the chosen algorithm by comparing the predicted *y* values with the true *y* values.

```python
X = data.drop('Y', axis=1)
y = data['Y']
```

### Feature selection

We will select only the features that show high correlations coefficients (r >= 0.15) with the target label.

```python
# define feature selection
features_to_remove = correlations.loc[:, correlations.loc['Y'] < 0.15].columns
data_sel = data.drop(features_to_remove, axis=1)

X_sel = data_sel.drop('Y', axis=1)
y = data_sel['Y']

X_selected_train, X_selected_test, y_train, y_test = train_test_split(X_sel, y, test_size=0.15, random_state=42)
```

```python
features_to_remove
```

```python
X_sel.shape
```

Now our dataset has only four features (and the target label *y*).

We will proceed to hyperparamenter tuning via GridSearch for each of the following models:
1. Logistic regression
2. KNN
3. DecisionTree
4. Random Forest
5. SVM

*Disclaimer*: XGBoost was also attempted, but it was eventually excluded due to its high computational cost (especially for the machine this project was worked on).

### Hyperparameter Turning via Grid Search

```python
models = ['LogisticRegression', 'DecisionTreeClassifier', 'RandomForestClassifier', 
          #'XGBClassifier', 
          'KNeighborsClassifier', 'SVC']
param_grid = dict().fromkeys(models)

param_grid['LogisticRegression'] = {'C': [1/0.0001, 1/0.001, 1/0.01, 1/0.1]} #inverse of regularization: 1/lambda

param_grid['DecisionTreeClassifier'] = {'criterion': ['entropy', 'log_loss', 'gini'], # measure of impurity
                                        'max_depth': np.arange(1, len(X_selected_train.columns), 1), # how many levels does the tree have?
                                        #'min_samples_split': [3, 4, 5], # how many samples are needed to make a split?
                                        'ccp_alpha': np.arange(0, 0.040, 0.005)# cost-complexity pruning
                                       }

param_grid['RandomForestClassifier'] = {'n_estimators': [30, 50, 75, 100, 150, 200], # number of trees in the forest
                                        'max_depth': np.arange(1, len(X_selected_train.columns), 1), # how many levels does the tree have?
                                        'min_samples_split': [3, 4, 5], # how many samples are needed to make a split?
                                        'warm_start': ['True', 'False']
                                       } 

#param_grid['XGBClassifier'] = {'max_depth': np.arange(1, len(X_selected_train.columns), 1), # how many levels does the tree have?
#                               'min_samples_split': [3, 4, 5], # how many samples are sufficient to make a split?
#                            }

param_grid['KNeighborsClassifier'] = {'n_neighbors' : np.arange(5, 35, 5)} # how many neighbors should be group together? 

param_grid['SVC'] = {'C': [0.0001, 0.001, 0.01, 0.1], # regularization term
                     'kernel': ['rbf', 'linear', 'poly'] # kernel trick method
                     }


def load_model(model):
    if model == 'LogisticRegression':
        return LogisticRegression()
    if model == "DecisionTreeClassifier":
        return DecisionTreeClassifier(random_state=42)
    if model == 'RandomForestClassifier':
        return RandomForestClassifier(random_state=42)
#    if model == 'XGBClassifier':
#        return XGBClassifier(random_state=42)
    if model == 'KNeighborsClassifier':
        return KNeighborsClassifier()
    if model == 'SVC':
        return SVC(random_state=42)

def confusion_matrix_plot(y_actual, y_pred, test_set=True):
    cm = confusion_matrix(y_actual, y_pred)
    plt.figure(figsize=(8, 5))
    sns.heatmap(cm, annot=True, fmt='.2f')
    plt.xlabel('Predicted')
    if test_set:
        plt.ylabel('Actual (from test set)')
    else: 
        plt.ylabel('Actual (from train set)')
    
dict_models = {}
for model in models:
    estimator = load_model(model)
    gs = GridSearchCV(estimator = estimator, param_grid = param_grid[model])
    gs.fit(X_selected_train, y_train)
    dict_models[model] = gs.best_estimator_
    y_pred_test = dict_models[model].predict(X_selected_test)
    y_pred_train = dict_models[model].predict(X_selected_train)
    print("\nMODEL REPORT: ", dict_models[model],
          "\n- against the test set:\n", classification_report(y_test, y_pred_test),
          "\n")
    confusion_matrix_plot(y_test, y_pred_test, test_set=True)
    plt.savefig(str(dict_models[model])+'_against_test.pdf')
    plt.show(), plt.close()
    print("\n- against the train set:\n", classification_report(y_train, y_pred_train),
          "\n")
    confusion_matrix_plot(y_train, y_pred_train, test_set=False)
    plt.savefig(str(dict_models[model])+'_against_train.pdf')
    plt.show(), plt.close()
```

Remarkably, the KKN model (with 20 neighbors) is 74% accurate. Notice that the F1-scores calculated with the target label of both the test and train sets are comparable (with only the F1-score calculated against the train set slightly below the 73% threshold). This suggests that the model is not overfitting, and generalizes fairly well. 

#### Post-pruning decision trees with cost complexity pruning

In addition to the KNeighbors Classifier, we ought to notice that the Decision Tree classifier performs just below the threshold against the train set, whereas accuracy against the test set is way below our threshold. These results may suggest that pre-pruning techniques to prevent overfitting (i.e., the hyperparameter tuning performed above via Grid Search) could not actually control for it. In what follows, we will therefore apply an alternative GridSearchCV procedure on the Decision Tree algorithm that will attempt to bypass overfitting by applying post-pruning via *cost complexity pruning*. 

Before doing any post-pruning technique, let's make sure that our tuned decision tree model actually benefit from all the features we selected above.

```python
feature_names = list(X_selected_train.columns)
feature_imp = dict_models['DecisionTreeClassifier'].feature_importances_
feature_imp_indices = np.argsort(feature_imp)

plt.figure(figsize=(8, 5))
plt.title("Feature importances")
plt.barh(range(len(feature_imp_indices)), feature_imp[feature_imp_indices])
plt.xlabel("importance")
plt.yticks(range(len(feature_imp_indices)), [feature_names[i] for i in feature_imp_indices])
plt.ylabel("feature")
plt.savefig('DecisionTree_FeatureImportance.pdf'), plt.show(), plt.close()
```

As feature *X5* has the lowest importance, we might try to remove it, and see if that help the algorithm to perform better with a properly grid-searched cost complexity pruning value.

```python
X_selected2_train = X_selected_train.drop(['X5'], axis=1)
X_selected2_test = X_selected_test.drop(['X5'], axis=1)

param_grid_dt_ccp = {'ccp_alpha': np.arange(0, 0.4, 0.005)}

gs_dt_ccp = GridSearchCV(estimator = DecisionTreeClassifier(random_state=42), param_grid = param_grid_dt_ccp)
gs_dt_ccp.fit(X_selected2_train, y_train)
dt_ccp = gs_dt_ccp.best_estimator_
y_pred_dt_test2 = dt_ccp.predict(X_selected2_test)
y_pred_dt_train2 = dt_ccp.predict(X_selected2_train)
print("\nMODEL REPORT: ", dt_ccp,
"\n- against the test set:\n", classification_report(y_test, y_pred_dt_test2),
"\n")
confusion_matrix_plot(y_test, y_pred_dt_test2, test_set=True)
plt.savefig('DecisionTree_select_against_test.pdf'), plt.show(), plt.close()
print("\n- against the train set:\n", classification_report(y_train, y_pred_dt_train2),
"\n")
confusion_matrix_plot(y_train, y_pred_dt_train2, test_set=False)
plt.savefig('DecisionTree_select_against_train.pdf'), plt.show(), plt.close()
```

Removing the less importance feature has definitely helped the model reaching the target accuracy (>= 73%) in both the test and train set, without even implementing any cost pruning complexity.
